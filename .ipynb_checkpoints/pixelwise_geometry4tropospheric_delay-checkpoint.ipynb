```python
import os
import numpy as np
from matplotlib import ticker, dates as mdates, pyplot as plt
from pysar.utils import readfile, writefile, ptime, plot as pp

work_dir = '/Users/yunjunz/insarlab/Galapagos/GalapagosSenDT128/PYSAR/PIXELWISE_GEOMETRY'
os.chdir(work_dir)
print('go to directory: '+work_dir)

inc_angle_file = './../INPUTS/geometryRadar.h5'
inc_angle, atr = readfile.read(inc_angle_file, datasetName='incidenceAngle')

vel_trop_diff_file = 'velocityEcmwf_diffInc.h5'
vel_trop_diff = readfile.read(vel_trop_diff_file)[0]


# save point timeseries of trop difference on volcanoes using cmd
# tsview.py ECMWF_diffInc.h5 --yx 690 1200 --nodisplay  # 1 - Wolf
# tsview.py ECMWF_diffInc.h5 --yx 1340 1360 --nodisplay # 2 - Fernandina
# tsview.py ECMWF_diffInc.h5 --yx 1910 410 --nodisplay  # 3 - Sierra Negra


date_list = list(np.loadtxt('y690_x1200_ts.txt', dtype=bytes).astype(str)[:, 0])
dates = ptime.date_list2vector(date_list)[0]

ts_1 = np.loadtxt('y690_x1200_ts.txt', dtype=bytes).astype(str)[:, 1].astype(np.float)
ts_2 = np.loadtxt('y1340_x1360_ts.txt', dtype=bytes).astype(str)[:, 1].astype(np.float)
ts_3 = np.loadtxt('y1910_x410_ts.txt', dtype=bytes).astype(str)[:, 1].astype(np.float)
```

```python
from matplotlib.ticker import MultipleLocator
font_size = 12
marker_size = 5

fig = plt.figure(figsize=[7, 3])
ax11 = fig.add_axes([0.00, 0.15, 0.20, 0.80])   # [x0, y0, width, height]
ax12 = fig.add_axes([0.21, 0.15, 0.20, 0.80])
ax13 = fig.add_axes([0.42, 0.15, 0.40, 0.80])
cax1 = fig.add_axes([0.025, 0.00, 0.15, 0.05])
cax2 = fig.add_axes([0.235, 0.00, 0.15, 0.05], facecolor='white', alpha=0.7)

#ax12 = ax11.twiny()
im1 = ax11.imshow(inc_angle, cmap='jet')
im2 = ax12.imshow(vel_trop_diff*1000., cmap='jet')
for ax in [ax11, ax12]:
    ax.invert_xaxis()
    ax.set_yticks(np.arange(0,2500,1000))
    ax.xaxis.tick_top()
    ax.tick_params(which='both', direction='in', labelsize=font_size, bottom=True, top=True, left=True, right=True)
ax12.set_yticklabels([])
ax11.set_ylabel('Azimuth (pix.)', fontsize=font_size)
pp.set_shared_xlabel([ax11, ax12], label='Range (pix.)', font_size=font_size)
#ax11.set_xlabel('Range (pix.)', fontsize=font_size)
ax11.xaxis.set_label_position("top")

cbar1 = fig.colorbar(im1, cax=cax1, orientation='horizontal')
cbar1.set_label('degree', fontsize=font_size)
cbar1.locator = ticker.MaxNLocator(nbins=3)
cbar1.update_ticks()

cbar2 = fig.colorbar(im2, cax=cax2, orientation='horizontal')
cbar2.set_label('mm/year', fontsize=font_size)
cbar2.locator = ticker.MaxNLocator(nbins=2)
cbar2.update_ticks()

for cbar in [cbar1, cbar2]:
    cbar.ax.xaxis.set_label_position('top')
for cax in [cax1, cax2]:
    cax.tick_params(direction='in', labelsize=font_size)

ax13.scatter(dates, ts_1*1000., s=marker_size**2, label='Wolf')
ax13.scatter(dates, ts_2*1000.+5, s=marker_size**2, label='Fernandina')
ax13.scatter(dates, ts_3*1000.+9, s=marker_size**2, label='Sierra Negra')
ax13.set_ylim([-4, 12])
ax13.set_ylabel('LOS Displacement (mm)')
ax13.yaxis.set_label_position("right")
ax13.yaxis.tick_right()
ax13.xaxis.set_major_locator(mdates.YearLocator(1))
ax13.tick_params(which='both', direction='in', labelsize=font_size, bottom=True, top=True, left=True, right=True)

#ax13.legend(loc='lower center', frameon=False, borderaxespad=0.1, fontsize=font_size*0.7, ncol=3)
ax13.text(0.05, 0.10, 'Wolf',         transform=ax13.transAxes, horizontalalignment='left', fontsize=font_size)
ax13.text(0.05, 0.35, 'Fernandina',   transform=ax13.transAxes, horizontalalignment='left', fontsize=font_size)
ax13.text(0.05, 0.70, 'Sierra Negra', transform=ax13.transAxes, horizontalalignment='left', fontsize=font_size)

plt.savefig('pixel_inc4trop.jpg', dpi=300)
plt.show()
```
