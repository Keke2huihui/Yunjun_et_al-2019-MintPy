```python
## Figure 4 - Galapagos DEM in Shaded Relief

import os
import matplotlib.pyplot as plt
from pysar.simulation.plot import *
from pysar.objects import sensor
from pysar.utils import readfile, plot as pp, utils as ut
from pysar import view

work_dir = os.path.expanduser('~/development/python/2018_ComputGeosci_PySAR')
os.chdir(work_dir)
print('Go to directory: '+work_dir)

geo_box = (-91.75, 0.2, -90.7, -1.1)  # WNES -91.75/-90.5/-1.15/0.3

proj_dir = os.path.expanduser('~/insarlab/Galapagos/GalapagosSenDT128')
dem_file = os.path.join(proj_dir, 'DEM/dem_masked.h5')
dem = pp.read_dem(dem_file, geo_box=geo_box)[0]
```

```python
print_msg=True
inps = pp.cmd_line_parse()
inps.disp_dem_contour = False

fig, ax = plt.subplots(figsize=[4, 6])
m = pp.BasemapExt(llcrnrlon=geo_box[0], llcrnrlat=geo_box[3],
                  urcrnrlon=geo_box[2], urcrnrlat=geo_box[1],
                  projection='cyl', resolution='i', suppress_ticks=False, ax=ax)
m = pp.plot_dem_background(ax=m, geo_box=geo_box, dem=dem, inps=inps, print_msg=print_msg)

m.draw_scale_bar(loc=[0.25, 0.8, 0.03], ax=ax, font_size=font_size)
m.draw_lalo_label(geo_box, ax=ax, lalo_step=0.5, labels=[1, 0, 0, 1], font_size=font_size, print_msg=print_msg)
plt.savefig('Galapagos_dem.png', bbox_inches='tight', transparent=True, dpi=fig_dpi)
plt.show()
```
