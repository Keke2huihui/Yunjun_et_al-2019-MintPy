```python
# Figure 12 - Automatic outlier detection based on timeseries phase residual
# with standardised residual of Median Absolute Deviation
import os
import numpy as np
import matplotlib.pyplot as plt
from pysar.simulation.plot import *
from pysar.objects import sensor
from pysar.timeseries_rms import plot_rms_bar

work_dir = os.path.expanduser('~/insarlab/PYSAR_EXAMPLE/SierraNegraAlosAT133/PYSAR')
os.chdir(work_dir)
print('Go to directory: '+work_dir)
proj_name = sensor.project_name2sensor_name(work_dir)[1]

# Opt 1 - plot
rms_txt_file = 'rms_timeseriesResidual_ramp.txt'
file_content = np.loadtxt(rms_txt_file, dtype=bytes).astype(str)
rms = file_content[:, 1].astype(np.float) * 1000.
date_list = list(file_content[:, 0])

fig, ax = plt.subplots(figsize=[4, 3])
ax = plot_rms_bar(ax, date_list, rms.tolist())
# save figure
out_file = 'rms_{}.jpg'.format(proj_name)
plt.savefig(out_file, bbox_inches='tight', transparent=True, dpi=fig_dpi)
print('saved to file:', out_file)

## Opt 2 - call timeseries_rms.py
#cmd = 'timeseries_rms.py timeseriesResidual.h5 -t pysarApp_template.txt'
#print(cmd)
#os.system(cmd)
#print('check saved figure file: rms_timeseriesResidual_ramp.pdf')

plt.show()
```
